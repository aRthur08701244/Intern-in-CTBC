```python
import os
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
import re
import ast
import json

def clearSentence(sentence):
    sentence = re.sub('-KY', '', sentence)
    sentence = re.sub('\*', '', sentence)
    return sentence

def svo2u(svo):
    (s, v, o) = svo
    svo = []
    for i in (s, v, o):
        svo.append(w2v_model.wv.get_vector(i, norm=True))
    return svo

def label(string):
    string = string.split('_')
    try:
        re = stockReturns[string[2]][string[0]]
    except:
        try:
            date = str(int(string[0])+1)
            re = stockReturns[string[2]][date]
        except:
            try:
                date = str(int(string[0])+2)
                re = stockReturns[string[2]][date]
            except:
                re = 0

    if re > 0:
        label = 1
    elif re < 0:
        label = -1
    else:
        label = 0

    return label
```
