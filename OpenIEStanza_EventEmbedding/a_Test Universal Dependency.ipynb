```python
EB = pd.DataFrame()

for file in ['20191', '20192', '20193', '20194', '20195', '20196', '20197', '20198', '20201', '20211']:
    print(file)
    news = pd.read_csv('./input/EB_SegAndRawNews/' + file + '_cnyesnews.csv')
    news['segmented'] = news['segmented'].apply(lambda x :ast.literal_eval(x))
    news = news[['date', 'all', 'segmented']]
    news['date'] = news.date.apply(lambda x : del_slash(x))
    
    print(news.shape[0])
    for i in news.index:
        print(i)
        if len(news.loc[i, 'segmented']) == 0:
            continue

        news_id = news.loc[i, 'date'] + '_' + str(i)
        EB = pd.concat([EB, pd.DataFrame({'News': [news_id], 'All': [news.loc[i, 'all']], 'SVO': [[]]})])

        Dep = nlp([news.loc[i, 'segmented']]) # Universal Dependency
        Dep = Dep.to_dict()
        Dep = pd.DataFrame(Dep[0])

        verbDF = Dep[Dep['upos'] == 'VERB']
        verbText = verbDF['text'].values
        verbID = verbDF['id'].values
        verbHead = verbDF['head'].values

        for (m, n) in zip(verbID, verbText):
            Verb = n
            Subject = Dep[(Dep['head'] == m) & (Dep['deprel'] == 'nsubj')]['text'].values
            Object = Dep[(Dep['head'] == m) & (Dep['deprel'] == 'obj')]['text'].values
            if (len(Subject) == 0) or (len(Object) == 0):
                continue
            for (o, p) in itertools.product(Subject, Object):
                SVO = [o, n, p]
    #             EB = pd.concat([EB, pd.DataFrame({'News': [news_id], 'Verb': [n], 'Subject': [o], 'Object': [p]})]) 較不好的方法
    #             print(EB.loc[EB['News'] == news_id, 'SVO'][0])
                EB.loc[EB['News'] == news_id, 'SVO'][0].append(SVO)

        if (i%500) == 0:
            EB.to_csv('./input/EB_EventExtraction/' + file + '_' + str(i//500) + '.csv')
            EB = pd.DataFrame()
        else:
            if i == news.index[-1]:
                EB.to_csv('./input/EB_EventExtraction/' + file + '_' + str(i//500+1) + '.csv')
                EB = pd.DataFrame()
```
