```python
marketIndex = stockPrice[stockPrice['證券代碼'] == 'Y9999  '].reset_index(drop=True)
```

```python
marketIndex
```

```python
marketTrend = {}
```

```python
for i in range(2013, 2022):
    for j in range(1, 13):
        if len(str(j)) == 1:
            start = '/'.join([str(i), '0'+str(j), '01'])
            end = '/'.join([str(i), '0'+str(j), '31'])
        else:
            start = '/'.join([str(i), str(j), '01'])
            end = '/'.join([str(i), str(j), '31'])
        
        monthlyIndex = marketIndex[marketIndex['年月日']>start].reset_index(drop=True)
        monthlyIndex = monthlyIndex[monthlyIndex['年月日']<end].reset_index(drop=True)
        
        n = monthlyIndex.shape[0]
        n_P = sum(monthlyIndex['報酬率％'].astype(float)>0)
        n_N = n - n_P
        ratio = n_P/n
        marketTrend[start[2:4]+start[5:7]] = ratio*100
        
#         print(monthlyIndex)
#         print(start, end)
```

```python
list(marketTrend.keys())
list(marketTrend.values())
marketTrend
```

```python
%matplotlib inline

fig = plt.figure(figsize=(30,4))
ax = plt.axes()
# ax.plot(return_df.columns, return_df.iloc[0, :], label='daily')
ax.plot(range(len(marketTrend.keys())), list(marketTrend.values()), label='Positive Ratio')
ax.plot(range(len(marketTrend.keys())), [50]*len(marketTrend.keys()))
# ax.plot(return_df.columns, return_df.iloc[1, :], label='0050_daily')
# ax.plot(return_df.index, return_df.iloc[:, 1]-1, label='0050_DailyReturn')
ax.set_xlabel('Time')  # Add an x-label to the axes.
ax.set_ylabel('Ratio(%)')  # Add a y-label to the axes.
ax.legend();  # Add a legend.
```
