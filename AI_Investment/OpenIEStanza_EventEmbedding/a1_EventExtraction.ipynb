```python
EB = pd.DataFrame()

import warnings
warnings.filterwarnings("ignore")
from wakepy import set_keepawake, unset_keepawake

set_keepawake(keep_screen_awake=True)

# for file in ['20191', '20192', '20193', '20194', '20195', '20196', '20197', '20198', '20201', '20211']:
# for file in ['20192', '20193', '20194', '20195', '20196', '20197', '20198', '20201', '20211']:
    
for i, j in itertools.product(range(2013, 2022), range(1, 13)):
    file = str(i)+str(j)
    print(file, end=': ')
    if file in ['20191', '20192', '20193', '20194', '20195', '20196', '20197', '20198', '20201', '20211']:
        print('Already Processed')
        continue
    news = pd.read_csv('./input/EB_SegAndRawNews/' + file + '_cnyesnews.csv')
    news = news[['date', 'all', 'segmented']]
    news['segmented'] = news['segmented'].apply(lambda x : ast.literal_eval(x))
    news['date'] = news.date.apply(lambda x : del_slash(x))
    news = news.dropna()
    news = news.reset_index(drop=True)
    
    print(news.shape[0])
    for i in sorted(news.index):

        news_id = news.loc[i, 'date'] + '_' + str(i)
        
        SVO = extractor.triples_main(ccT2S.convert(news.loc[i, 'all']))
#         SVO
        
        for j in range(len(SVO)):
            for k in range(len(SVO[j])):
                SVO[j][k] = ccS2T.convert(SVO[j][k])
        
        EB = pd.concat([EB, pd.DataFrame({'News': [news_id], 'All': [news.loc[i, 'all']], 'SVO': [SVO]})])
        
        if ((i+1)%100 == 0):
            EB.to_csv('./input/EB_EventExtraction2/' + file + '_' + str((i+1)//100) + '.csv')
            print('Export', './input/EB_EventExtraction2/' + file + '_' + str((i+1)//100) + '.csv')
            EB = pd.DataFrame()
        else:
            if i == news.index[-1]:
                EB.to_csv('./input/EB_EventExtraction2/' + file + '_' + str((i+1)//100+1) + '.csv')
                print('Export', './input/EB_EventExtraction2/' + file + '_' + str((i+1)//100+1) + '.csv')
                EB = pd.DataFrame()

warnings.filterwarnings("default")

set_keepawake(keep_screen_awake=False)
```
