```python
marketTrend = {}
```

```python
for i in range(2013, 2022):
    for j in range(1, 13):
        if len(str(j)) == 1:
            start = '/'.join([str(i), '0'+str(j), '01'])
            end = '/'.join([str(i), '0'+str(j), '31'])
        else:
            start = '/'.join([str(i), str(j), '01'])
            end = '/'.join([str(i), str(j), '31'])
        
        monthlyIndex = marketIndex[marketIndex['']>start].reset_index(drop=True)
        monthlyIndex = monthlyIndex[monthlyIndex['']<end].reset_index(drop=True)
        
        n = monthlyIndex.shape[0]
        n_P = sum(monthlyIndex[''].astype(float)>0)
        n_N = n - n_P
        ratio = n_P/n
        marketTrend[start[2:4]+start[5:7]] = ratio*100
        
#         print(monthlyIndex)
#         print(start, end)
```

```python
def RiseOrFall(com, date):
    try:
        re = stockReturn[com][date]
        if re > 0:
            re = 1
        elif re == 0:
            re = 0
        elif re < 0:
            re = -1
    except:
        re = 0
    return re
```

```python
with open('./input/stockReturns.json') as data_file:
    stockReturn = json.load(data_file)
```

```python
Ratio = {}

loc = './input/WB_TransNews/train/'
for i in range(2013, 2020):
    print(i)
    for j in range(1, 13):
        news = pd.read_csv(loc+str(i)+str(j)+'_cnyesnews.csv')
        news.date = news.date.astype(str)
        returns = np.array([])
        
        for k in news.index:
            returns = np.append(returns, RiseOrFall(news.loc[k, 'com'], news.loc[k, 'date']))

        Ratio[str(i)+str(j)] = sum(returns==1)*100 / (sum(returns==1)+sum(returns==-1))
```
